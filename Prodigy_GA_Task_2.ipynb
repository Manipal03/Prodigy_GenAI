```python
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion pipeline
model_id = "CompVis/stable-diffusion-v1-4"
device = "cuda" if torch.cuda.is_available() else "cpu"
pipeline = StableDiffusionPipeline.from_pretrained(model_id, use_auth_token=True)
pipeline.to(device)

def generate_stable_diffusion_image(prompt):
    # Generate image
    image = pipeline(prompt).images[0]
    return image

# Example usage
prompt = "A beautiful sunset over the mountains"
generated_image = generate_stable_diffusion_image(prompt)

# Save or display the image
generated_image.save("stable_diffusion_generated_image.png")
```
